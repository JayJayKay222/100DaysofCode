### Higher Lower Game 

```python
from art import logo, vs
from game_data import data
import random
import os 


# a function to get a random account 
def get_account():
    return random.choice(data)
    
# store the account data
def store_account(account):
    name = account["name"]
    description = account["description"]
    country = account["country"]
    return f"{name}, a {description}, from {country}."
    

# compare accounts for numnber of followers
def compare_account(answer, follower_A, follower_B):
    if answer == "A":
        if follower_A > follower_B:
            return answer == "A"
    else:
        return answer == "B"

# create the game
def game():
    trials = 0
    game_Continue = True
    account_B = get_account()
    
    while game_Continue:
        account_A = account_B
        account_B = get_account()
        while account_A == account_B:
            account_B = get_account()
            
        # print logo of the game
        print(logo)
        # compare the account followers     
        print(f"Compare: {store_account(account_A)}")
        print(vs)      
        print(f"Against: {store_account(account_B)}")

        # ask user for a guess
        user_guess = input("Who has more followers? Type 'A' or 'B': ").upper()   
        
        while user_guess not in ['A', 'B']:
            user_guess = input("Who has more followers? Type 'A' or 'B': ").upper()   
            
        is_True = compare_account(user_guess, account_A["follower_count"], account_B["follower_count"])
        if is_True:
            trials += 1
            print(f"You're right! Current score: {trials}.\n")
            os.system("clear")
        else:
            game_Continue = False
            print(f"Sorry that's wrong. Final score: {trials}")
game()    
# make the game repeatable
```
