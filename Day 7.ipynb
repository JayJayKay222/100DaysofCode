```python
"""
1. Randomly choose a word 
2. Prompt user to guess a letter and assign to avariable called guess
3. Check if the guessed letter is in the chosen_word
4. Repplace the blanks with the guessed word
5. While loop to let the user guess again until the user wins
"""
```

```python
import random

from hangman_words import word_list
from hangman_art import stages, logo

chosen_word = random.choice(word_list)

print(logo)

display = []
for letter in range(len(chosen_word)):
    display += "_"

GameEnd = False
life = 6

while not GameEnd:
    guess = input("Guess a letter: ").lower()
    
    if guess in display:
                print(f"You've already guessed {guess}.\n")
    for i in range(len(chosen_word)):
        letter = chosen_word[i]
        
        if letter == guess:
            display[i] = letter
            
    if guess not in chosen_word:
        life -= 1
        print(f"You guessed {guess}, that's not in the word.\n")
        print(stages[-life - 2])
        
        if life == 0:
            GameEnd = True
            print("You lost!")
            print(f"The correct word is {chosen_word}.\nKeep up you can do better next time.")
        
    print(f"{''.join(display)}")
    
    if "_" not in display:
        GameEnd = True
        print("You Won!")


    
```
